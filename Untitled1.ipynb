imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

reading dataset and cleaning


```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/MyDrive/datasets/Cancer_Data.csv")
print(df.head())

df.info()
df.describe()

df = df.drop(['id','Unnamed: 32'], axis=1)


df.head()

X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

```

Missing values

diagnose count

```python
sns.countplot(x='diagnosis', data=df, palette=['green','red'])
plt.title("Count of Benign (B) vs Malignant (M)")
plt.xlabel("Diagnosis")
plt.ylabel("Number of Patients")
plt.show()
```

train and scale


```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

Build train and make predictions


```python
model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

```

Evaluation


```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
